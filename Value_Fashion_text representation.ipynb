```python
#Create a fit_cv function used to build the Bag-of-Words vectorizer with the corpus
def fit_cv(post_corpus):
  cv_vect = CountVectorizer(tokenizer=lambda x: x,
                            preprocessor=lambda x: x)
  cv_vect.fit(post_corpus)
  return cv_vect
```
